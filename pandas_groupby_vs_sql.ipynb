```python
import pandas as pd

raw = {
    'name':['Brandon', 'Rohan', 'Kathy', 'Han', 'Jill', 'Ken', 'Young', 'Tray', 'Derek', 'Janet', 'Jack'],
    'country':['Korea', 'Japan', 'Japan', 'China', 'USA', 'USA', 'Singapore', 'Korea', 'Singapore', 'China', 'Singapore'],
    'gender':['M', 'F', 'F', 'F', 'M', 'F', 'M', 'M', 'M', 'M', 'M'],
    'score':[60, 90, 99, 45, 56, 78, 95, 54, 61, 64, 54],
    'age':[25, 34, 26, 65, 19, 32, 54, 21, 18, 56, 78]
}

student = pd.DataFrame(raw)
student
```

```python
# SELECT country, mean(score) AS mean_score, max(score) AS max_score, mean(age) AS mean_age, min(age) AS min_age
# FROM student
# GROUPBY country
```

## 1. Aggregate per eachÂ column

```python

# Make empty dataframe
student_agg = pd.DataFrame()

# Groupby and aggregate per column
student_group = student.groupby('country')
student_agg['avg_score'] = student_group['score'].mean()
student_agg['max_score'] = student_group['score'].max()
student_agg['avg_age'] = student_group['age'].mean()
student_agg['min_age'] = student_group['age'].min()


student_agg = student_agg.reset_index()
student_agg
```

## 2. Aggregate by DataFrame.agg method

```python
# Make empty dataframe
student_agg = pd.DataFrame()

# Add new columns by using .agg method
student_agg[['avg_score', 'max_score']] = student_group['score'].agg(['mean','max'])
student_agg[['avg_age', 'min_age']] = student_group['age'].agg(['mean','min'])
student_agg.head()
```

## 3. Aggregate by dictionary and DataFrame.agg

```python
# Make empty dataframe
student_agg = pd.DataFrame()


# Using dictionary to represent all the aggregation
agg_dict = {
    'score':['mean','max'],
    'age':['mean','min']
}

# Aggregate all at once
student_agg = student_group.agg(agg_dict)
student_agg

```

```python
# Replace multi-index to singular index
student_agg.columns = [('_').join(column) for column in student_agg.columns]
student_agg
```
